# **Alura Store Latam**

## Importación de datos



```python
import pandas as pd

url = "https://raw.githubusercontent.com/alura-es-cursos/challenge1-data-science-latam/refs/heads/main/base-de-datos-challenge1-latam/tienda_1%20.csv"
url2 = "https://raw.githubusercontent.com/alura-es-cursos/challenge1-data-science-latam/refs/heads/main/base-de-datos-challenge1-latam/tienda_2.csv"
url3 = "https://raw.githubusercontent.com/alura-es-cursos/challenge1-data-science-latam/refs/heads/main/base-de-datos-challenge1-latam/tienda_3.csv"
url4 = "https://raw.githubusercontent.com/alura-es-cursos/challenge1-data-science-latam/refs/heads/main/base-de-datos-challenge1-latam/tienda_4.csv"

tienda = pd.read_csv(url)
tienda2 = pd.read_csv(url2)
tienda3 = pd.read_csv(url3)
tienda4 = pd.read_csv(url4)

tienda.head()
```

```python
tienda4.info()
```

```python
tienda.sort_index(ascending=False)
```

```python
tienda.isnull().values.any()
```

##1. Análisis de facturación



### **Ingreso total por cada tienda**

```python
ingreso_tienda1 = tienda ['Precio'].sum()
ingreso_tienda2 = tienda2['Precio'].sum()
ingreso_tienda3 = tienda3['Precio'].sum()
ingreso_tienda4 = tienda4['Precio'].sum()
```

```python
ingresos_totales = pd.DataFrame({
    'Tienda': ['Tienda 1', 'Tienda 2', 'Tienda 3',  'Tienda 4'],
    'Ingreso_total': [ingreso_tienda1, ingreso_tienda2, ingreso_tienda3, ingreso_tienda4]
})

pd.options.display.float_format = '{:,.0f}'.format
ingresos_totales
```

```python
import matplotlib.pyplot as plt

plt.figure(figsize=(8,5))
plt.bar(ingresos_totales['Tienda'], ingresos_totales['Ingreso_total'], color='mediumseagreen')

plt.title('Ingreso total por tienda', fontsize=14, fontweight='bold')
plt.xlabel('Tienda')
plt.ylabel('Ingreso total ($)')
plt.xticks(rotation=0)
plt.tight_layout()
plt.show()
```

## 2. Ventas por categoría

### Cantidad de productos vendidos por categoría en cada tienda.

```python
cat_t1 = tienda.groupby('Categoría del Producto')['Producto'].count().sort_values(ascending=False)
cat_t2 = tienda2.groupby('Categoría del Producto')['Producto'].count().sort_values(ascending=False)
cat_t3 = tienda3.groupby('Categoría del Producto')['Producto'].count().sort_values(ascending=False)
cat_t4 = tienda4.groupby('Categoría del Producto')['Producto'].count().sort_values(ascending=False)
```

```python
df_categorias = pd.DataFrame({
    'Tienda 1': cat_t1,
    'Tienda 2': cat_t2,
    'Tienda 3': cat_t3,
    'Tienda 4': cat_t4
}).fillna(0).astype(int)

df_categorias
```

```python
mas_popular = df_categorias.idxmax()
mas_popular
```

```python
plt.figure(figsize=(7,7))
plt.pie(cat_t1,
        labels=cat_t1.index,
        autopct='%1.1f%%',
        startangle=90,
        colors=plt.cm.Paired.colors)

plt.title('Distribución de categorías más vendidas – Tienda 1', fontsize=14, fontweight='bold')
plt.tight_layout()
plt.show()
```

```python
colores = ['#4CAF50', '#2196F3', '#FF9800', '#9C27B0']

df_categorias.plot(
    kind='barh',
    figsize=(10,5),
    color=colores
)

plt.title('Cantidad de productos vendidos por categoría en cada tienda', fontsize=14, fontweight='bold')
plt.xlabel('Cantidad de productos vendidos')
plt.ylabel('Categoría')
plt.legend(title='Tiendas', loc='best')
plt.tight_layout()
plt.show()
```

## 3. Calificación promedio de la tienda


```python
valoracion_t1 = tienda ['Calificación'].mean()
valoracion_t2 = tienda2['Calificación'].mean()
valoracion_t3 = tienda3['Calificación'].mean()
valoracion_t4 = tienda4['Calificación'].mean()
```

```python
df_calificacion = pd.DataFrame({
    'Tienda' : ['Tienda 1', 'Tienda 2', 'Tienda 3', 'Tienda 4'],
    'Calificación_promedio' : [valoracion_t1, valoracion_t2, valoracion_t3, valoracion_t4]
})

pd.options.display.float_format = '{:.2f}'.format
df_calificacion.sort_values(by='Calificación_promedio', ascending=False)
```

```python
df_calificacion.plot(kind='bar', x='Tienda', y='Calificación_promedio',
                     color=plt.cm.Set2.colors, figsize=(8,5), legend=False)

plt.title('Calificación promedio por tienda')
plt.ylabel('Promedio de calificación')
plt.xlabel('Tiendas')
plt.ylim(0, 5)
plt.xticks(rotation=0)
plt.tight_layout()
plt.show()
```

## 5. Envío promedio por tienda

```python
costo_env_t1 = tienda ['Costo de envío'].mean()
costo_env_t2 = tienda2['Costo de envío'].mean()
costo_env_t3 = tienda3['Costo de envío'].mean()
costo_env_t4 = tienda4['Costo de envío'].mean()
```

```python
df_costo_promedio = pd.DataFrame({
    'Tienda': ['Tienda 1', 'Tienda 2', 'Tienda 3', 'Tienda 4'],
    'Costo de envío promedio': [costo_env_t1, costo_env_t2, costo_env_t3, costo_env_t4]
})

pd.options.display.float_format = '{:.2f}'.format
df_costo_promedio
```

```python
df_costo_promedio.plot(kind='bar', x='Tienda', y='Costo de envío promedio',
                       color=plt.cm.Set2.colors, legend=False, figsize=(8,5))

plt.title('Costo de envío promedio por tienda')
plt.xlabel('Tienda')
plt.ylabel('Costo de envío (promedio)')
plt.tight_layout()
plt.show()
```
