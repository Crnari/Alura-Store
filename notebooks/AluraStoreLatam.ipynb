# **Analisis exploratorio de los datos de Alura Store**

## Importación de datos



```python
import pandas as pd

url = "https://raw.githubusercontent.com/alura-es-cursos/challenge1-data-science-latam/refs/heads/main/base-de-datos-challenge1-latam/tienda_1%20.csv"
url2 = "https://raw.githubusercontent.com/alura-es-cursos/challenge1-data-science-latam/refs/heads/main/base-de-datos-challenge1-latam/tienda_2.csv"
url3 = "https://raw.githubusercontent.com/alura-es-cursos/challenge1-data-science-latam/refs/heads/main/base-de-datos-challenge1-latam/tienda_3.csv"
url4 = "https://raw.githubusercontent.com/alura-es-cursos/challenge1-data-science-latam/refs/heads/main/base-de-datos-challenge1-latam/tienda_4.csv"

tienda = pd.read_csv(url)
tienda2 = pd.read_csv(url2)
tienda3 = pd.read_csv(url3)
tienda4 = pd.read_csv(url4)

tienda.head()
```

##1. Análisis de facturación



### **Ingreso total por cada tienda**

```python
ingreso_tienda1 = tienda ['Precio'].sum()
ingreso_tienda2 = tienda2['Precio'].sum()
ingreso_tienda3 = tienda3['Precio'].sum()
ingreso_tienda4 = tienda4['Precio'].sum()
```

```python
ingresos_totales = pd.DataFrame({
    'Tienda': ['Tienda 1', 'Tienda 2', 'Tienda 3',  'Tienda 4'],
    'Ingreso_total': [ingreso_tienda1, ingreso_tienda2, ingreso_tienda3, ingreso_tienda4]
})

pd.options.display.float_format = '{:,.0f}'.format
ingresos_totales
```

## 2. Ventas por categoría

### Cantidad de productos vendidos por categoría en cada tienda.

```python
cat_t1 = tienda.groupby('Categoría del Producto')['Producto'].count().sort_values(ascending=False)
cat_t2 = tienda2.groupby('Categoría del Producto')['Producto'].count().sort_values(ascending=False)
cat_t3 = tienda3.groupby('Categoría del Producto')['Producto'].count().sort_values(ascending=False)
cat_t4 = tienda4.groupby('Categoría del Producto')['Producto'].count().sort_values(ascending=False)
```

```python
df_categorias = pd.DataFrame({
    'Tienda 1': cat_t1,
    'Tienda 2': cat_t2,
    'Tienda 3': cat_t3,
    'Tienda 4': cat_t4
}).fillna(0).astype(int)

df_categorias
```

## 3. Calificación promedio de la tienda


```python
valoracion_t1 = tienda ['Calificación'].mean()
valoracion_t2 = tienda2['Calificación'].mean()
valoracion_t3 = tienda3['Calificación'].mean()
valoracion_t4 = tienda4['Calificación'].mean()
```

# **Analisis de rendimiento de Alura Store**

## Introduccion

El Sr. Juan desea iniciar un nuevo emprendimiento y, para financiarlo, debe decidir cuál de las tiendas de su cadena Alura Store es la menos eficiente y por lo tanto la mejor candidata para vender.
Con este propósito, se realizó un análisis de datos utilizando herramientas de ciencia de datos como Pandas y Matplotlib, evaluando aspectos clave del rendimiento de cada tienda.

Este informe sintetiza los hallazgos del análisis, incluyendo métricas como:

* Ingresos totales por tienda

* Categorías de productos más y menos vendidas

* Calificaciones promedio de los clientes

* Productos más y menos vendidos

* Coste de envío promedio

Con base en los datos, se ofrece una recomendación final fundamentada, identificando la tienda menos eficiente.

## Desarrollo del Análisis

### Ingresos Totales por Tienda

Los ingresos totales representan el desempeño económico global de cada tienda. Al sumar las ventas registradas, se obtuvo:

```python
import matplotlib.pyplot as plt

colores = ['#4CAF50', '#2196F3', '#FFC107', '#FF5722']

plt.figure(figsize=(9,5))
plt.bar(
    ingresos_totales['Tienda'],
    ingresos_totales['Ingreso_total'],
    color=colores
)

plt.title('Ingreso total por tienda', fontsize=14, fontweight='bold')
plt.xlabel('Tienda')
plt.ylabel('Ingreso total ($)')
plt.xticks(rotation=0)

for i, valor in enumerate(ingresos_totales['Ingreso_total']):
    plt.text(
        i,
        valor + (valor * 0.01),
        f"${valor:,.0f}",
        ha='center',
        fontsize=10,
        fontweight='bold'
    )

plt.tight_layout()
plt.show()
```

* Tienda 4 es la que menos ingresos genera (1,038,375,700).

* Esto la posiciona como la menor contribuyente económica dentro de Alura Store.

* Vender esta tienda causaría el menor impacto financiero a la cadena.

## Categorías de Productos Más Vendidas

Se evaluó cuántos productos se vendieron por categoría en cada tienda.

```python
df_categorias.plot(kind='bar', figsize=(14,7))
plt.title("Productos vendidos por categoría en cada tienda")
plt.xlabel("Categoría de producto")
plt.ylabel("Cantidad vendida")
plt.xticks(rotation=45)
plt.tight_layout()
plt.show()
```

* Artículos para el hogar, Deportes y diversión, Libros, Juguetes, y Muebles presentan una demanda considerablemente alta en todas las tiendas.

* Las categorías con menor volumen de ventas tienden a ser más especializadas o con menor rotación natural, lo cual es esperable en un entorno de ventas diversificado.

* Aunque existen diferencias entre tiendas, la distribución por categorías muestra que todas ellas cubren necesidades similares y comparten un patrón estable de consumo por parte de los clientes.

## Calificacion promedio de los clientes

Las reseñas permiten evaluar la experiencia del cliente en cada tienda.

```python
colores = ['#4CAF50', '#2196F3', '#FFC107', '#FF5722']

df_calificacion.plot(kind="bar", x="Tienda", y="Calificación_promedio", figsize=(8,5), color= colores, legend = False)
plt.title("Calificación Promedio por Tienda")
plt.ylabel("Calificación")
plt.ylim(3.8, 4.1)
plt.xticks(rotation=0)
plt.tight_layout()
plt.show()
```

* Las puntuaciones están cercanas a 4.0, lo que refleja un buen nivel de aceptación del servicio y los productos ofrecidos.

* Aunque algunas tiendas presentan valores ligeramente superiores o inferiores, estas diferencias no son lo suficientemente amplias como para señalar una disparidad significativa en la calidad del servicio.

* El comportamiento homogéneo de las reseñas sugiere que la política de atención, la gestión de productos y el estándar de experiencia del cliente se mantiene equilibrado entre todas las tiendas.

## Productos mas y menos vendidos

Los productos mas vendidos de cada una de las tiendas son las siguientes:

```python
producto_top_t1 = tienda ['Producto'].value_counts().idxmax()
producto_top_t2 = tienda2['Producto'].value_counts().idxmax()
producto_top_t3 = tienda3['Producto'].value_counts().idxmax()
producto_top_t4 = tienda4['Producto'].value_counts().idxmax()

df_top_productos = pd.DataFrame({
    'Tienda': ['Tienda 1', 'Tienda 2', 'Tienda 3', 'Tienda 4'],
    'Producto más vendido': [
        producto_top_t1, producto_top_t2, producto_top_t3, producto_top_t4
    ]
})

df_top_productos
```

Los productos menos vendidos de cada tienda son:

```python
producto_menosv_t1 = tienda ['Producto'].value_counts().idxmin()
producto_menosv_t2 = tienda2['Producto'].value_counts().idxmin()
producto_menosv_t3 = tienda3['Producto'].value_counts().idxmin()
producto_menosv_t4 = tienda4['Producto'].value_counts().idxmin()

df_menos_vendidos = pd.DataFrame({
    'Tienda': ['Tienda 1', 'Tienda 2', 'Tienda 3', 'Tienda4'],
    'Producto menos vendido': [
        producto_menosv_t1, producto_menosv_t2, producto_menosv_t3, producto_menosv_t4
    ]
})

df_menos_vendidos
```

* Los productos más vendidos suelen pertenecer a categorías de consumo frecuente o aquellos que ofrecen utilidad práctica en el hogar o en entornos cotidianos.

* Los productos menos vendidos suelen estar asociados a nichos específicos o preferencias más particulares del cliente, lo que explica su menor salida.

* Este comportamiento es esperado en una cadena con un catálogo amplio, ya que no todos los productos tienen el mismo nivel de rotación ni el mismo público objetivo.

## Costo de envio promedio

Los costos logísticos pueden influir directamente en la conversión de ventas.

```python
df_costo_promedio.plot(kind="bar", x="Tienda", y="Costo de envío promedio", figsize=(8,5), color= colores, legend = False)
plt.title("Costo de Envío Promedio por Tienda")
plt.ylabel("Costo de Envío")
plt.xticks(rotation=0)
plt.tight_layout()
plt.show()
```

* La tienda 1 presenta un mayor costo de envio promedialmente, esto puede deberse a la cantidad de productos vendidos

* La tienda 2 y 3 se mantienen casi que en un mismo nivel en cuanto al costo de envio

* La tienda 4 es la que presenta un menor costo de envio, puede deberse a al poco nivel de demanda que hay en la mismo

## Conclusion y recomendacion final

Luego de analizar los ingresos, categorías de productos, reseñas, productos más vendidos y costos de envío, se concluye que la tienda que el Sr. Juan debería vender es:

**La tienda 4**

* Es la tienda con menores ingresos totales, lo que indica el peor desempeño financiero.

* Tiene buenos niveles de ventas por categoría, pero no los convierte en rentabilidad.

* Su calificación es aceptable, pero inferior a las tiendas de mejor rendimiento.

* A pesar de tener el mejor costo de envío, el impacto positivo no se refleja en los indicadores.

* Vender cualquiera de las otras tiendas causaría un impacto económico mucho mayor en la cadena.


**Conclusión final:**

La Tienda 4 es la menos eficiente globalmente, por lo que venderla representa el menor riesgo y la mayor conveniencia para el Sr. Juan al iniciar su nuevo emprendimiento.
